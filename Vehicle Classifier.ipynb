```python
from sklearn import tree
```

```python

# features etanol=1 gasolina = 0
# lables car= 0 e caminhao= 1


#PARTE 1 DO MACHINE LEARNING : COLLECT TRAINING DATA
features = [[40, 1],[35,0],[400, 0],[380,1]]
labels = [0, 0, 1, 1]

#PARTE 2 DO MACHINE LEARNING : TRAIN CLASSIFIER - DECISION TREE

clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)  
#find patterns in data
pred = clf.predict([[250,1]])
```
